```python
import pandas as pd
from tabulate import tabulate

def load_data():
    
    multi = pd.read_csv("../data/multi_errors.csv", sep = "\t")
    EWR_single = pd.read_csv("../data/EWR_single_errors.csv", sep = "\t")
    JFK_single = pd.read_csv("../data/JFK_single_errors.csv", sep = "\t")
    LGA_single = pd.read_csv("../data/LGA_single_errors.csv", sep = "\t")

    cnn_errors = pd.read_csv("../data/cnn_errors.csv")
    cnn_errors = cnn_errors.drop(cnn_errors.columns[[0]], axis=1)
    cnn_lstm_errors = pd.read_csv("../data/cnn_+_lstm_errors.csv")
    cnn_lstm_errors = cnn_lstm_errors.drop(cnn_lstm_errors.columns[[0]], axis=1)
    lstm_errors = pd.read_csv("../data/lstm_errors.csv")
    lstm_errors = lstm_errors.drop(lstm_errors.columns[[0]], axis=1)
    
    print "## La Guardia ##"
    LGA = pd.DataFrame()
    LGA["multi"] = multi["LGA"].values
    LGA["single"] = LGA_single.values
    print tabulate(LGA, headers='keys', tablefmt='psql')
    
    print "## EWR ##"
    EWR = pd.DataFrame()
    EWR["multi"] = multi["EWR"].values
    EWR["single"] = EWR_single.values
    print tabulate(EWR, headers='keys', tablefmt='psql')
    
    print "## JFK ##"
    JFK = pd.DataFrame()
    JFK["multi"] = multi["JFK"].values
    JFK["single"] = JFK_single.values
    print tabulate(JFK, headers='keys', tablefmt='psql')
    
    print " ## CNN network ##"
    print tabulate(cnn_errors, headers='keys', tablefmt='psql')
    
    print " ## CNN + LSTM network ##"
    print tabulate(cnn_lstm_errors, headers='keys', tablefmt='psql')
    
    print " ## LSTM network ##"
    print tabulate(lstm_errors, headers='keys', tablefmt='psql')
    
    
def main():
    load_data()
    
    
if __name__ == "__main__":
    main()
```
